# Model Inference

## Import Library

```python
# import library
from tensorflow import keras
import pandas as pd
import numpy  as np
import joblib
from tensorflow.keras.models import load_model
```

# Data Loading

```python
inf = pd.read_csv('inf.csv')
inf
```

```python
inf_final = inf.drop('churn_risk_score', axis=1)
```

```python
inf_final
```

# Model Loading

```python
#load pipeline
pipeline = joblib.load(open('pipeline.pkl', 'rb'))
```

```python
# load model
model = load_model('modelf_imp_wq5.h5')
```

```python
dfinalnew = pipeline.transform(inf_final)
dfinalnew
```

```python
# predict result
predictions = model.predict(dfinalnew)
predictions = np.where(predictions>0.5,1,0)
predictions
```

```python
# show data infer + prediction
inf_final['churn_risk_score'] = predictions
inf_final
```
